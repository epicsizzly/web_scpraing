```python
from bs4 import BeautifulSoup

# Read the index.html file
with open('laptop.html', 'r', encoding='utf-8') as file:
    html_content = file.read()

# Parse the HTML content
soup = BeautifulSoup(html_content, 'html.parser')

# Find all divs with the specified class
all_items = soup.find_all('div', class_='sg-col-20-of-24 s-result-item s-asin sg-col-0-of-12 sg-col-16-of-20 AdHolder sg-col s-widget-spacing-small sg-col-12-of-16')

# Initialize an empty list to store extracted data
data_list = []

# Iterate through each div
for item in all_items:
    # Find span for title
    title_span = item.find('span', class_='a-size-medium a-color-base a-text-normal')
    title = title_span.text.strip() if title_span else ''

    # Find span for price
    price_span = item.find('span', class_='a-price-whole')
    price = price_span.text.strip() if price_span else ''

    # Find span for rating
    rating_span = item.find('span', class_='a-icon a-icon-star-small a-star-small-4-5 aok-align-bottom')
    rating = rating_span.text.strip() if rating_span else ''

    # Append data to the list
    data_list.append({'title': title, 'price': price, 'rating': rating})

# Write data to data.json
import json

with open('data.json', 'w', encoding='utf-8') as json_file:
    json.dump(data_list, json_file, indent=4)

```
